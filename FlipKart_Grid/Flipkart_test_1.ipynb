```
import pandas as pd
import numpy as np
```

```
train_pd = pd.read_csv('drive/My Drive/FlipKart_Grid/training.csv')
test_pd = pd.read_csv('drive/My Drive/FlipKart_Grid/test.csv')
```

```
from keras.applications.vgg16 import VGG16
from keras.optimizers import SGD, Adam
from keras.layers import GlobalAveragePooling2D
from keras import Model
```

```
from keras.applications.imagenet_utils import preprocess_input
from keras.preprocessing.image import img_to_array,load_img
```

```
Y_train = []
for i in train_pd.index:
  y = train_pd.iloc[i]
  y_t = [i,y.x1, y.x2, y.y1, y.y2]
  Y_train.append(y_t)
Y_train = pd.DataFrame(Y_train, columns =['Index','X1','X2','Y1','Y2'],index = train_pd.index)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Y_train['Index'],Y_train[['X1','X2','Y1','Y2']], test_size=0.2)
```

```
test_data = []
for i in x_test.index:
  imgPath = 'drive/My Drive/FlipKart_Grid/train/' + train_pd.loc[i].image_name
  imgTransit = load_img(imgPath)
  imgArray = img_to_array(imgTransit)
  imgPre = preprocess_input(imgArray)
  test_data.append(imgPre)
```

```
from keras import applications
from keras.applications.imagenet_utils import preprocess_input
from keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
from keras import optimizers
from keras.models import Sequential, Model 
from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D
from keras import backend as k 
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, TensorBoard, EarlyStopping
from keras.layers import Conv2D, MaxPooling2D


vgg_layer="block2_pool"

from keras.applications.vgg16 import VGG16
base_model = VGG16(weights='imagenet', include_top=False,input_shape=[480, 640, 3])
for layer in base_model.layers:
  layer.trainable = False
  if layer.name==vgg_layer: break
vgg_half =base_model.get_layer(vgg_layer).output

x=Conv2D(64,(3, 3), activation='relu', padding='same',name='conv1')(vgg_half)
x=Flatten(name='flatten1')(x)
x=Dense(32,name="dense1")(x)
x=Dense(4,name="output")(x)
model=Model(inputs=base_model.input, outputs=x)

  
print(model.summary())
```

```
model.compile(loss='mean_squared_error',optimizer='sgd',metrics=['mae'])
```

```
callbacks = [EarlyStopping(monitor='val_loss', patience=2),ModelCheckpoint(filepath='drive/My Drive/FlipKart_Grid/best_model.h5', monitor='val_loss', save_best_only=True)]
```

```
x1_data_test = list(x_test.X1)
x2_data_test = list(x_test.X2)
y1_data_test = list(x_test.Y1)
y2_data_test = list(x_test.Y2)
data_test = []
for u in range(len(x_test.index)):
  data_test.append([x1_data_test[u],x2_data_test[u],y1_data_test[u],y2_data_test])
data_test = np.asarray(data_test)
```

```
for i in range(0,len(x_train.index),10):
  x_data = x_train[i:10+i]
  y_data = y_train[i:10+i]
  x1_data = list(x_data.X1)
  x2_data = list(x_data.X2)
  y1_data = list(x_data.Y1)
  y2_data = list(x_data.Y2)
  
  input_data = []
  for t in range(10):
    input_data.append([x1_data[t],x2_data,y1_data[t],y2_data[t]])
  input_data = np.asarray(input_data)
  data = []
  for e in range(10):
    #######################################################  Image Loading  ####################################################
    imgPath_train = 'drive/My Drive/FlipKart_Grid/train/' + train_pd.loc[i+e].image_name
    imgTransit_train = load_img(imgPath_train)
    imgArray_train = img_to_array(imgTransit_train)
    imgPre_train = preprocess_input(imgArray_train)
    data.append(imgPre_train)
  data = np.asarray(data)
  training = model.fit(data,input_data,epochs = 1,callbacks = callbacks,verbose=1,batch_size=10,validation_data=(test_data,data_test))
```
